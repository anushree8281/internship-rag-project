```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
sentences = [
    "Artificial Intelligence is transforming technology.",
    "Machine Learning is a branch of AI.",
    "Deep Learning uses neural networks.",
    "Dogs are friendly animals.",
    "Cats like sleeping.",
    "Kerala is famous for backwaters.",
    "Python is a programming language.",
    "Football is a popular sport.",
    "RAG improves answer accuracy.",
    "Vector databases store embeddings."
]
```

```python
import time

embeddings = []

for sentence in sentences:
    success = False

    while not success:
        try:
            response = client.models.embed_content(
                model="models/gemini-embedding-001",
                contents=sentence
            )

            embeddings.append(response.embeddings[0].values)
            success = True

        except Exception as e:
            print("Quota exceeded. Waiting 30 seconds...")
            time.sleep(30)
print("Embeddings created:", len(embeddings))
```

```python
query = "Tell me about Artificial Intelligence"
```

```python
try:
    query_response = client.models.embed_content(
        model="models/gemini-embedding-001",
        contents=query
    )

    query_embedding = query_response.embeddings[0].values

except Exception as e:
    print("Error generating query embedding")
    print("Reason:", e)
```

```python
try:
    similarities = cosine_similarity(
        [query_embedding],
        embeddings
    )[0]

except Exception as e:
    print("Error calculating similarity")
    print("Reason:", e)
```

```python
try:
    top_indices = np.argsort(similarities)[::-1][:5]

    for i in top_indices:
        print()
        print("Sentence:", sentences[i])
        print("Similarity:", round(similarities[i], 4))

except Exception as e:
    print("Error retrieving top results")
    print("Reason:", e)
```
